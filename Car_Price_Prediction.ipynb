```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('car data.csv', )
df.head()
```

```python
df.info()
```

```python
df.drop('Car_Name', axis=1, inplace=True)
df = pd.get_dummies(df, columns=['Fuel_Type', 'Seller_Type', 'Transmission'], dtype=float)
df
```

```python
df.shape
```

```python
X = df.drop('Present_Price', axis=1)
y = df['Present_Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# scaler
scaler = MinMaxScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train))
X_test = pd.DataFrame(scaler.transform(X_test))
```

```python
X_train
```

```python
# linear regression gradient descent
# N - number of iterations
# lr - learning rate

class Gradient():
  def fit(self, X_train, y_train, N=1000, lr=0.2):
    self.X_train=X_train
    self.y_train=y_train.to_numpy().reshape(len(y_train), 1)
    self.m, self.n = X_train.shape
    self.N=N
    self.lr=lr
    self.w = np.zeros((self.n+1, 1))
    ones = pd.Series(np.ones((self.m)), name='ones')
    self.X = pd.concat([ones.reset_index(drop=True), self.X_train.reset_index(drop=True)], axis=1)

    for i in range(self.N):
      # loss = (1/(2*self.m))*(self.y_train - self.X @ self.w)**2
      grad =  (1/(self.m)) * self.X.T @ (self.X @ self.w - self.y_train)
      self.w -= self.lr * grad
      # print(grad)


  def predict(self, X_test):
    self.X_test = X_test
    ones = pd.Series(np.ones((self.X_test.shape[0])), name='ones')
    self.y_pred = pd.concat([ones.reset_index(drop=True), self.X_test.reset_index(drop=True)], axis=1) @ self.w
    # print(self.y_pred)
    return self.y_pred

  def score(self, X_test, y_test):
    self.y_test = y_test.to_numpy().reshape(len(y_test), 1)
    y_pred = self.predict(X_test)
    s1 = ((self.y_test - y_pred)**2).sum() # Sum of Squared Residuals
    s2 = ((self.y_test - self.y_test.mean())**2).sum() # Total Sum of Squares
    r2 = np.round((1 - s1/s2)*100, 2)
    # print(y_pred)
    # print(self.y_test)
    return f'Implemented Models R2 Score is: {float(r2[0])}%'


model = Gradient()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# ridge regression gradient descent
# N - number of iterations
# lr - learning rate
# alpha - regularization parameter

class Ridge():
  def fit(self, X_train, y_train, N=1000, lr=0.2, alpha=0.0001):
    self.X_train=X_train
    self.y_train=y_train.to_numpy().reshape(len(y_train), 1)
    self.m, self.n = X_train.shape
    self.N=N
    self.lr=lr
    self.alpha=alpha
    self.w = np.zeros((self.n+1, 1))
    ones = pd.Series(np.ones((self.m)), name='ones')
    self.X = pd.concat([ones.reset_index(drop=True), self.X_train.reset_index(drop=True)], axis=1)

    for i in range(self.N):
      # loss = (1/(2*self.m))*(self.y_train - self.X @ self.w)**2
      grad =  (1/(self.m)) * self.X.T @ (self.X @ self.w - self.y_train) + 2*self.alpha * self.w
      self.w -= self.lr * grad
      # print(grad)


  def predict(self, X_test):
    self.X_test = X_test
    ones = pd.Series(np.ones((self.X_test.shape[0])), name='ones')
    self.y_pred = pd.concat([ones.reset_index(drop=True), self.X_test.reset_index(drop=True)], axis=1) @ self.w
    # print(self.y_pred)
    return self.y_pred

  def score(self, X_test, y_test):
    self.y_test = y_test.to_numpy().reshape(len(y_test), 1)
    y_pred = self.predict(X_test)
    s1 = ((self.y_test - y_pred)**2).sum() # Sum of Squared Residuals
    s2 = ((self.y_test - self.y_test.mean())**2).sum() # Total Sum of Squares
    r2 = np.round((1 - s1/s2)*100, 2)
    # print(y_pred)
    # print(self.y_test)
    return f'Implemented Models R2 Score is: {float(r2[0])}%'


model = Ridge()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
